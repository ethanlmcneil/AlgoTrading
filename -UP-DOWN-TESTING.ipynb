```python
import yfinance as yf
```

```python

data = yf.download(
    tickers = 'AAPL',
    start="2020-01-01", 
    end="2024-04-30", 
    interval = "1d",
    ignore_tz=True,
    auto_adjust=True, 
)


data['target next day % change'] = data['Close'].pct_change()*100

data.head()
```

```python
corr = data.corr()

corr
```

```python
data = data.drop(['Open', 'Low', 'High'], axis=1)
```

```python
data
```

```python
# find other features that are correlated with the target RSI


def calculate_rsi(data, window=14):
    # Calculate daily price changes
    delta = data['Close'].diff()
    
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    
    avg_gain = gain.rolling(window=window, min_periods=1).mean()
    avg_loss = loss.rolling(window=window, min_periods=1).mean()
    
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    
    return rsi



data['RSI'] = calculate_rsi(data)

```
